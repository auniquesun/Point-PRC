```python
import os
import h5py
```

```python
# --- step 1: read `obj_only`, `obj_bg` or `hardest` version data

# so_test_f = os.path.join(sonn_dir, 'obj_only', 'test_objectdataset.h5')
# so_test_f = os.path.join(sonn_dir, 'obj_bg', 'test_objectdataset.h5')
so_test_f = os.path.join(sonn_dir, 'hardest', 'test_objectdataset_augmentedrot_scale75.h5')

so_test_data = h5py.File(so_test_f)
so_test_pts = so_test_data['data'][:]
so_test_labels = so_test_data['label'][:]

print(so_test_pts.shape)
print(so_test_labels.shape)
```

```python
# --- step 2: map class names to indexes for sonn
cls2idx = {}
idx2cls = {}
with open(os.path.join(sonn_dir, 'shape_names.txt')) as fin:
    lines = fin.readlines()
    for i, line in enumerate(lines):
        if line.strip() != '':
            cls2idx[line.strip()] = i
            idx2cls[i] = line.strip()
```

```python
# --- step 3: filter out desired classes according to `so_11_classes`
selected_idx = []

for key in so_11_classes:
    if key in cls2idx:
        selected_idx.append(cls2idx[key])

assert len(selected_idx) == len(so_11_classes), f'{len(selected_idx)} < {len(so_11_classes)}, so there are lost classes!'
```

```python
# --- step 4: write desired pts and labels into .npy file
import numpy as np

# dst_dir = '../data/xset/sim2real/so_obj_only_11'
# dst_dir = '../data/xset/sim2real/so_obj_bg_11'
dst_dir = '../data/xset/sim2real/so_hardest_11'
cls2pts = {}

for pts, label in zip(so_test_pts, so_test_labels):
    if label in selected_idx:
        class_name = idx2cls[label]
        if class_name not in cls2pts:
            cls2pts[class_name] = [pts]
        else:
            cls2pts[class_name].append(pts)

for cls in cls2pts:
    save_dir = os.path.join(dst_dir, cls, 'test')

    if not os.path.exists(save_dir):    # make dir recursively if it doesn't exist
        os.makedirs(save_dir)
    
        class_points = cls2pts[cls]
        for i, pts in enumerate(class_points):
            idx = str(i).zfill(4)
            np.save(f'{save_dir}/{cls}_{idx}.npy', pts)
```

```python
# --- step 5: check the saved data
fname = 'cabinet/test/cabinet_0010.npy'
target_f = os.path.join(dst_dir, fname)
target_data = np.load(target_f)
print("target_data.shape:", target_data.shape)
print("first point:", target_data[0])
```
